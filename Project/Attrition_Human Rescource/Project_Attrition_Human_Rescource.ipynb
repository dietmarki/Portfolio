# alle features

```python
# train-/test-daten:
from sklearn.model_selection import train_test_split
```

```python
X01 = df02.drop("Attrition",axis=1)
y01 = df02["Attrition"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X01, y01, test_size=0.20, random_state=33)
```

```python
# alle X-daten standardisieren:
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python

# algorithmen: LogReg,KNeighbor(elbow),SVC(c/gamma-->gridsearch),naiveBayes,randomRorest
```

```python
# algorithmen importieren:
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

#initialisieren:
log = LogisticRegression()
knc = KNeighborsClassifier()
nab = GaussianNB()
rfc = RandomForestClassifier()
svc = SVC()
```

```python
# elbow method:
fehler = []
for k in range(1,10):
    knc = KNeighborsClassifier(n_neighbors=k)
    knc.fit(X_train,y_train)
    predictions_k = knc.predict(X_test)
    fehler.append(np.mean(predictions_k != y_test))
plt.plot(range(1,10),fehler)
#                               wählen  --> k = 2!  ---> k = 6 am besten!!
#plt.savefig(r"C:\Users\dietm\Documents\Data Science_Didi\12. Woche\Projekt\attrition\Plots\Plot_12.png")
```

```python
#beste werte für C und gamma (SVC,hyperebenen):
hyper_para = {"C":[0.1,1,10,100,1000],"gamma":[1,0.1,0.01,0.001,0.0001]}
# initialisieren von Gridsearch
grid = GridSearchCV(SVC(),hyper_para,refit=True)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# neu initialisieren nach "elbow" und "C/gamma" und trainieren:
log = LogisticRegression().fit(X_train,y_train)
knc = KNeighborsClassifier(n_neighbors=6).fit(X_train,y_train)
nab = GaussianNB().fit(X_train,y_train)
rfc = RandomForestClassifier(n_estimators=1000,random_state=33).fit(X_train,y_train)
svc = SVC(C=1000,gamma=0.0001).fit(X_train,y_train)
```

```python
# vorhersagen:
pred_log = log.predict(X_test)
pred_knc = knc.predict(X_test)
pred_nab = nab.predict(X_test)
pred_rfc = rfc.predict(X_test)
pred_svc = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,mean_absolute_error,classification_report
```

```python
print("LogReg      : {:.2f} %".format(accuracy_score(y_test,pred_log)*100))
print("KNeighbor   : {:.2f} %".format(accuracy_score(y_test,pred_knc)*100))
print("NaiveBayes  : {:.2f} %".format(accuracy_score(y_test,pred_nab)*100))
print("RandomForest: {:.2f} %".format(accuracy_score(y_test,pred_rfc)*100))
print("SVC         : {:.2f} %".format(accuracy_score(y_test,pred_svc)*100))
```

# reports

```python
#models in liste speichern:
models = [log,knc,nab,rfc,svc]
model_names = ["LogisticRegression","KNeighborsClassifier","NaiveBayes","RandomForestClassifier","SVC"]
# über for schleife werte in listen speichern:
accuracy_scores = []
precision_scores = []
recall_scores = []
error_rate = []


for i,m in enumerate(models):
    y_pred = m.predict(X_test)
    conf_matrix = confusion_matrix(y_test,y_pred)
    
    print(model_names[i])
    print(classification_report(y_test,y_pred))
    
    true_positiv = conf_matrix[0,0]
    true_negativ = conf_matrix[1,1]
    false_positiv = conf_matrix[0,1]
    false_negativ = conf_matrix[1,0]
    
    # werte: "accuracy","precision","recall","error"
    total = true_positiv + true_negativ + false_positiv + false_negativ
    accuracy = (true_positiv + true_negativ) / total
    precision = true_positiv / (true_positiv + false_positiv)
    recall = true_positiv / (true_positiv + false_negativ)
    error = (false_positiv + false_negativ) / total
    
    accuracy_scores.append(accuracy.round(2))
    precision_scores.append(precision.round(2))
    recall_scores.append(recall.round(2))
    error_rate.append(error.round(2))

# tabelle erzeugen mit den Listen der einzelnen werte:
ergebnisse_df = pd.DataFrame(list(zip(accuracy_scores,precision_scores,recall_scores,error_rate)),
                            index=model_names,columns=["accuracy","precision","recall","error"])

ergebnisse_df
```

```python
fig,axis = plt.subplots(1,5,figsize=(13,2))

sns.heatmap(confusion_matrix(y_test,pred_log),cbar = False,cmap="Blues_r",annot=True,fmt=".0f",ax=axis[0])
axis[0].set_title("LogReg")
sns.heatmap(confusion_matrix(y_test,pred_knc),cbar = False,cmap="Blues_r",annot=True,fmt=".0f",ax=axis[1])
axis[1].set_title("KNeighb")
sns.heatmap(confusion_matrix(y_test,pred_nab),cbar = False,cmap="Blues_r",annot=True,fmt=".0f",ax=axis[2])
axis[2].set_title("NaiBay")
sns.heatmap(confusion_matrix(y_test,pred_rfc),cbar = False,cmap="Blues_r",annot=True,fmt=".0f",ax=axis[3])
axis[3].set_title("RanFor")
sns.heatmap(confusion_matrix(y_test,pred_svc),cbar = False,cmap="Blues_r",annot=True,fmt=".0f",ax=axis[4])
axis[4].set_title("SVC")

#plt.savefig(r"C:\Users\dietm\Documents\Data Science_Didi\12. Woche\Projekt\attrition\Plots\Plot_02.png")
```

```python
# datensätze
X05 = x_pca
y05 = df02["Attrition"]
```

```python
# train_test_split
X_train05, X_test05, y_train05, y_test05 = train_test_split(X05, y05, test_size=0.20, random_state=33)
```

```python
#initialisieren:
log05 = LogisticRegression()
knc05 = KNeighborsClassifier()
nab05 = GaussianNB()
rfc05 = RandomForestClassifier()
svc05 = SVC()
```

```python
# elbow method:
fehler05 = []
for k in range(1,10):
    knc05 = KNeighborsClassifier(n_neighbors=k)
    knc05.fit(X_train05,y_train05)
    predictions_k05 = knc05.predict(X_test05)
    fehler05.append(np.mean(predictions_k05 != y_test05))
plt.plot(range(1,10),fehler05)
#                               wählen  --> k = 2!  ---> k = 6 am besten!!
#plt.savefig(r"C:\Users\dietm\Documents\Data Science_Didi\12. Woche\Projekt\attrition\Plots\Plot_15.png")
```

```python
#beste werte für C und gamma (SVC,hyperebenen):
hyper_para05 = {"C":[0.1,1,10,100,1000],"gamma":[1,0.1,0.01,0.001,0.0001]}
# initialisieren von Gridsearch
grid05 = GridSearchCV(SVC(),hyper_para05,refit=True)
grid05.fit(X_train05,y_train05)
```

```python
grid05.best_params_
```

```python
# neu initialisieren nach "elbow" und "C/gamma" und trainieren:
log05 = LogisticRegression().fit(X_train05,y_train05)
knc05 = KNeighborsClassifier(n_neighbors=6).fit(X_train05,y_train05)
nab05 = GaussianNB().fit(X_train05,y_train05)
rfc05 = RandomForestClassifier(n_estimators=1000,random_state=33).fit(X_train05,y_train05)
svc05 = SVC(C=100,gamma=1).fit(X_train05,y_train05)
```

```python
# vorhersagen:
pred_log05 = log05.predict(X_test05)
pred_knc05 = knc05.predict(X_test05)
pred_nab05 = nab05.predict(X_test05)
pred_rfc05 = rfc05.predict(X_test05)
pred_svc05 = svc05.predict(X_test05)
```

```python
# k_mean importieren:
from sklearn.cluster import KMeans
```

```python
# elbow method mit datensatz reduziert "x_pca" (1 dimension):
#             --> inertias (distanzen bzw. fehler)
kmeans_k = [KMeans(n_clusters=k, random_state=33).fit(x_pca) for k in range(1, 11)]
inertias = [model.inertia_ for model in kmeans_k]
```

```python
# visualisieren:
plt.grid()
plt.plot(range(1, 11),inertias,"bo-")
#plt.savefig(r"C:\Users\dietm\Documents\Data Science_Didi\12. Woche\Projekt\attrition\Plots\Plot_16.png")
```
